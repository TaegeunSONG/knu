```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import datasets
```

```python
(x_train, y_train), (x_test, y_test) = datasets.mnist.load_data()
```

```python
print(x_train.shape, x_test.shape)
```

```python
i=2460
print(y_train[i])
plt.imshow(x_train[i])
plt.colorbar()
plt.show()
# print(x_train[20])
```

```python
idx=[]

for id in range(10):
    for i in range(y_test.shape[0]):
        if y_test[i] == id:
          idx.append(i)
          break

fig, axes = plt.subplots(ncols=5, nrows=2, figsize=(18,6))
fig.subplots_adjust(hspace=0.4, wspace=0.4)
# plot first few images

for i in range(10):
	  axes[i//5, i%5].imshow(x_test[idx[i]], cmap=plt.get_cmap('gray'))
fig.show()
```

# Preprocessing (data shape)

```python
from tensorflow.keras.utils import to_categorical

cnn_data = x_train
cnn_label = to_categorical(y_train)
cnn_data_test = x_test
cnn_label_test = to_categorical(y_test)
```

```python
i=59999
print(y_train[i], cnn_label[i])
plt.imshow(cnn_data[i])
plt.show()
```

#Build model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten
from tensorflow.keras.utils import plot_model

model = Sequential()

model.add( Conv2D(filters=2, kernel_size=10, strides=2, input_shape=(28,28,1), \
                  activation='relu', dilation_rate=1, padding='same'))
model.add( MaxPooling2D(2))
model.add(Flatten())

model.add(Dense(10, activation='softmax'))

model.summary()
```

# Compile and Excute the model 

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics='accuracy')

history = model.fit(cnn_data, cnn_label, epochs=10, batch_size=100, validation_split=0.3)
```

# Testing and Learning curves

```python
scores = model.evaluate(cnn_data_test, cnn_label_test)

print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))

fig, axes = plt.subplots(ncols=2, nrows=1, figsize=(12,6))
fig.suptitle('Learning Curves')

axes[0].plot(history.history['loss'], 'o-', label='train')
axes[0].plot(history.history['val_loss'], 'o-', label='val')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Cross Entropy')
axes[0].set_title('Loss')
axes[0].legend()

axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('accuracy')
axes[1].plot(history.history['accuracy'], 'o-', label='train')
axes[1].plot(history.history['val_accuracy'], 'o-', label='val')
axes[1].set_title('accuracy')
axes[1].legend()

plt.show()
```

# Advance(!?) options

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping


cb_checkpoint = ModelCheckpoint(filepath='/content/cnn.hdf5', monitor='val_accuracy', mode = 'max',\
                                verbose=1, save_best_only=True)
cb_early_stopping = EarlyStopping(monitor='val_accuracy', mode = 'max', patience=2)


model = Sequential()
model.add( Conv2D(filters=2, kernel_size=10, strides=2, input_shape=(28,28,1), \
                  activation='relu', dilation_rate=1, padding='same'))
model.add( MaxPooling2D(2))
model.add(Flatten())

model.add(Dense(10, activation='softmax'))

model.summary()



model.compile(loss='binary_crossentropy', optimizer='adam', metrics='accuracy')

history = model.fit(cnn_data, cnn_label, epochs=10, batch_size=100, validation_split=0.3, \
                    callbacks=[cb_checkpoint, cb_early_stopping])
```

```python
from tensorflow.keras.models import load_model

model_call = load_model('/content/cnn.hdf5')
model_call.summary()
scores = model_call.evaluate(cnn_data_test, cnn_label_test)
```

# Tip to make deep 

```python
from tensorflow.keras.layers import Input

model = Sequential()


model.add(Input(shape=(28,28,1)))

for i in range(3):
  model.add( Conv2D(filters=2, kernel_size=10, strides=1, \
                    activation='relu', dilation_rate=1, padding='same'))
  model.add( MaxPooling2D(2))

model.add(Flatten())
model.add(Dense(10, activation='softmax'))

model.summary()
```
